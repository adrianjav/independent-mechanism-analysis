### Nonlinear mixings

We first define three different mixing functions

```python
import numpy as np
```

```python
def f_1(x, y):
    "Moderately nonlinear mixing"
    f0 = np.tanh(4*x - 2) + x + y/2
    f1 = np.tanh(4*y - 2) + y + x/2
    return f0, f1

def f_2(x, y):
    "Rather nonlinear mixing"
    f0 = np.tanh(y)/2 + x + y**2/2
    f1 = x**3 - x + np.tanh(y)
    return f0, f1

def f_3(x, y):
    "Non-bijective nonlinear mixing"
    f0 = y**3 + x
    f1 = np.tanh(y) + x**3
    return f0, f1
```

Replicating figures 1, 2, 3 in the paper:

Images of a grid in the square $[0,1] \times [0,1]$ using the three functions.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

def plot_grid(x,y, ax=None, **kwargs):
    ax = ax or plt.gca()
    segs1 = np.stack((x,y), axis=2)
    segs2 = segs1.transpose(1,0,2)
    ax.add_collection(LineCollection(segs1, **kwargs))
    ax.add_collection(LineCollection(segs2, **kwargs))
    ax.autoscale()
```

```python
fig, ax = plt.subplots()

grid_x,grid_y = np.meshgrid(np.linspace(0,1,20),np.linspace(0,1,20))
plot_grid(grid_x,grid_y, ax=ax,  color="lightgrey")

distx, disty = f_1(grid_x,grid_y)
plot_grid(distx, disty, ax=ax, color="C0")

plt.show()
```

```python
fig, ax = plt.subplots()

grid_x,grid_y = np.meshgrid(np.linspace(0,1,20),np.linspace(0,1,20))
plot_grid(grid_x,grid_y, ax=ax,  color="lightgrey")

distx, disty = f_2(grid_x,grid_y)
plot_grid(distx, disty, ax=ax, color="C0")

plt.show()
```

```python
fig, ax = plt.subplots()

grid_x,grid_y = np.meshgrid(np.linspace(0,1,20),np.linspace(0,1,20))
plot_grid(grid_x,grid_y, ax=ax,  color="lightgrey")

distx, disty = f_3(grid_x,grid_y)
plot_grid(distx, disty, ax=ax, color="C0")

plt.show()
```

### The Darmois construction

Maybe a simple way is to use Kernel Density Estimation

https://www.statsmodels.org/stable/nonparametric.html

Which exists both in a univariate (or multivariate) unconditional
https://www.statsmodels.org/stable/generated/statsmodels.nonparametric.kernel_density.KDEMultivariate.html#statsmodels.nonparametric.kernel_density.KDEMultivariate

or conditional fashion
https://www.statsmodels.org/stable/generated/statsmodels.nonparametric.kernel_density.KDEMultivariateConditional.html#statsmodels.nonparametric.kernel_density.KDEMultivariateConditional

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
nobs = 300
np.random.seed(1234)  # Seed random generator
# data = np.random.normal(size=nobs)
data = np.random.uniform(0.0,1.0, size=2000)
# dens = sm.nonparametric.KDEUnivariate(data)
dens = sm.nonparametric.KDEUnivariate(data)
dens.fit()
plt.plot(dens.cdf)
plt.show()

dens.cdf
```

```python
import statsmodels.api as sm
nobs = 300
c1 = np.random.normal(size=(nobs,1))
c2 = np.random.normal(2,1,size=(nobs,1))
dens_c = sm.nonparametric.KDEMultivariateConditional(endog=[c1],
    exog=[c2], dep_type='c', indep_type='c', bw='normal_reference')
dens_c.bw   # show computed bandwidth
```

```python
dens_c.cdf
```
